Read excel

```python
from holidays import BusCalendar
cal = BusCalendar()

def isColumn(colname):
    return not colname.startswith("Unnamed:")

with pd.ExcelFile("backlog.xlsx") as xls:
    df_read = pd.read_excel(xls, skiprows=6, skipfooter=1, usecols = isColumn)#, names = col_names, dtype={'creation_dt':'datetime64[ns]', 'completion_dt':'datetime64[ns]'}, date_format='%d.%m.%y')
df_read.columns = col_names

df = df_read.copy()
df[['creation_dt', 'completion_dt']] = df[['creation_dt', 'completion_dt']].apply(pd.to_datetime, format='%d.%m.%Y')
df['creation_dt'] = df['creation_dt'].fillna(datetime(2023, 1, 1))
df['completion_dt'] = df['completion_dt'].fillna(default_completion_dt)
df['project'] = df['project'].fillna("ND")
df['completion_status'] = df['completion_status'].fillna("ND")
df['min_effort'] = df[['preplanned_effort', 'planned_effort', 'fact_effort']].min(axis=1)
df['preplanned_effort'] = df['preplanned_effort'].fillna(1)
df['planned_effort'] = df['planned_effort'].fillna(1)
df['fact_effort'] = df['fact_effort'].fillna(1)
df['max_effort'] = df[['preplanned_effort', 'planned_effort', 'fact_effort']].max(axis=1)
df['duration'] = df.apply(lambda row: cal.date_diff(row['creation_dt'], row['completion_dt']), axis=1)

df_ini = df.copy()
```

Filter dataframe and print stats

```python
def print_stats(name, df):
        print(f"{name}\n\tcount:\t{df.duration.count()}\n\tduration average:\t{round(df.duration.mean())}\n\tduration median: \t{round(df.duration.median())}\n\teffort average:\t\t{round(df.planned_effort.mean())}\n\teffort median: \t\t{round(df.planned_effort.median())}")
    

start_dt = datetime(2022, 6, 1)
finish_dt = datetime(2023, 2, 1)
EFFORT = 'max_effort'


print(f"Uncleaned:\n{df_ini[['duration', 'max_effort', 'min_effort']].describe()}")


print(f"\nStats for projects: {projects_lst}\n and statuses: {statuses_lst}\n\n")
df = df_ini[(df_ini['duration'] >= 5)
            & (df_ini[EFFORT] >= 8)
            & (df_ini['project'].isin(projects_lst))
            & (df_ini['completion_status'].isin(statuses_lst))
            #& (df_ini['priority_grp'] != 2)
            ]

print(f"Unfiltered:\n{df[['duration', 'max_effort', 'min_effort']].describe()}")
#df = df[(df['duration'] >= 10) & (df['duration'] <= 200)]


df_filtered = df[(df['creation_dt'] >= start_dt) & 
        (df['completion_dt'] >= finish_dt)]

print(f"\nFilters applied: creation date>={start_dt}; completion date >={finish_dt}\n")

print(f"Full set:\n{df_filtered[['duration', 'max_effort', 'min_effort']].describe()}")

df_8_40 = df_filtered[(df_filtered[EFFORT] >= 0)
                      & (df_filtered[EFFORT] < 40)
                      #& (df_filtered['duration'] >= 21)
                      #& (df_filtered['duration'] <= 90)
                      ]

print(f"\n8-40:\n{df_8_40[['duration', 'max_effort', 'min_effort']].describe()}")
            
df_40_80 = df_filtered[(df_filtered[EFFORT] >= 40)
                       & (df_filtered[EFFORT] < 80)
                       #& (df_filtered['duration'] >= 50)
                       #& (df_filtered['duration'] <= 169)
                       ]
print(f"\n40-80:\n{df_40_80[['duration', 'max_effort', 'min_effort']].describe()}")

df_80_200 = df_filtered[(df_filtered[EFFORT] >= 80)
                        & (df_filtered[EFFORT] < 200)
                        #& (df_filtered['duration'] >= 76)
                        #& (df_filtered['duration'] <= 212)
                        ]
print(f"\n80-200:\n{df_80_200[['duration', 'max_effort', 'min_effort']].describe()}")

```
